```python
import pandas as pd

car_data_raw = pd.read_csv("https://cdn.c18l.org/vehicles_lab.csv")
```

```python
cars = car_data_raw.loc[:,["price","year","condition","fuel","odometer","transmission", "manufacturer"]]
cars.head()
#These are the varibles I assumed woudl best shape my model
```

```python
cars["good vehicles"]= (
    (cars['condition'] == "good")
).astype(int)
```

```python
cars = cars.loc[cars['year'] > 1965]
cars = cars.loc[cars['year'] < 2022]
# Since the range of years for cars were so limitless, decided to go 65 and up
```

```python
cars['transmission'].unique()
```

```python
cars.head()
```

```python
cars = cars.dropna()
```

```python
cars['car_age'] = 2022 - (cars['year'])
cars['car_age'].describe()
```

```python
cars=pd.get_dummies(cars, columns=['condition','transmission','fuel' ], dummy_na=True)
#dummy coding each of these dichotomous variables and categorical varibles where it would allow a value system to describe each column
```

```python
cars.describe()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(
    cars,
    train_size=0.8,
    random_state=42)

model = LogisticRegression(multi_class='multinomial').fit(
    X=train_data.loc[:, [
        'odometer', 'car_age',
        'good vehicles','transmission_automatic','transmission_manual',
        'fuel_gas','fuel_hybrid', 'fuel_hybrid'
        ]],
    y=train_data['price']
)
## Tried running this code a few times but it would never run properly. The ram crashed each time. I got stuck here
```
