## Feature Engineering

To help, I've included a column_builder() utility function that will create a new boolean column based on whether a string of text appears in any of (1) the recipe title; (2) the recipe description; or (3) the recipe tags.

The code below I turned into my own using one of the categories listed and placing in key words found in the description that I thought would also matter. When looking for a recipe typically you want things that are easy and depending on the time of day or time of year, it matters what you're cooking. So I added Breakfast, Lunch, Dinner, kid-friendly, and party categories to become boolean columns.

```python
def column_builder(category: str, recipes: pd.DataFrame) -> pd.DataFrame:
    recipes[f'is_{category}'] = ((
        recipes['categories'].str.contains(f'{category}', na=False, case=False)
    ) | (
        recipes['title'].str.contains(f'{category}', na=False, case=False)
    )).astype(int)

    return recipes


categories = [
    'easy',
    'breakfast',
    'dinner',
    'lunch',
    'kid-friendly',
    'party'
    # Add any additional keywords here
]

for category in categories:
    recipes = column_builder(category, recipes)

recipes['is_easy'].describe()
```
